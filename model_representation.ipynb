```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
x_values = np.array([2104, 1600, 2400, 1416, 3000])
y_values = np.array([400, 330, 369, 232, 540])

# Reshape x_values to a 2D array, required by scikit-learn
x_values = x_values.reshape(-1, 1)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(x_values, y_values)

# Make predictions
x_pred = np.linspace(min(x_values), max(x_values), 100).reshape(-1, 1)
y_pred = model.predict(x_pred)

# Plot the data and the linear regression line
plt.scatter(x_values, y_values, label='Data points')
plt.plot(x_pred, y_pred, color='red', label='Linear Regression Model')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Model')
plt.legend()
plt.show()

# Display the coefficients of the linear regression model
print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_[0]}')
```
